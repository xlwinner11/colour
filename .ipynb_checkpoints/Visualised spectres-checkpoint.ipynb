```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
spectres_filename = "/Users/glpopov/Downloads/MacbethColorChecker.xls"
spectres_file = pd.ExcelFile(spectres_filename)

sheet = spectres_file.parse('Sheet1')
print(sheet)
```

```python
wavelengths = list(range(380, 785, 5))
print(wavelengths)
```

```python
spectres = dict()
for i in range(1, 25):
    spectres[i] = list()
    current_column = sheet['Unnamed: ' + str(i)]
    for j, value in enumerate(current_column):
        if j <= 1:
            continue
        spectres[i].append(value)
```

```python
print(spectres)
```

```python
fig, ax = plt.subplots(nrows=3, ncols=8, figsize=(60, 40))

for i, row in enumerate(ax):
    for j, col in enumerate(row):
        colour_number = 8 * i + j + 1
        col.plot(wavelengths, spectres[colour_number])
        ax[i,j].set_ylim(top=1, bottom=0)
        ax[i,j].set_title('Colour # ' + str(colour_number))
        

plt.show()
```

```python
import colour
import colour.colorimetry as clr
import colour.plotting as clplt

swatches = []

# for i in range(3):
#     for j in range(8):
colour_number = 8 * 0 + 3 + 1
sd = spectres[colour_number]
spd = dict()
for k, value in enumerate(sd):
    spd[wavelengths[k]] = value
spd = colour.SpectralDistribution(spd)
csd = clr.sd_to_XYZ(spd)
csd = colour.XYZ_to_sRGB(csd / 100)
for k, v in enumerate(csd):
    if v < 0:
        csd[k] = 0
    if v > 1:
        csd[k] = 1
swatches.append(clplt.ColourSwatch(RGB=csd))
clplt.plot_multi_colour_swatches(swatches)
```

```python
reflectances = []
start = 380
step = 4
wavelengths = list(range(380, 784, 4))
```

```python
with open('reflect_db.reflect') as fp:
    line = fp.readline()
    lines = 0
    empty = 0
    while line:
        if line[0] != '#' and line[0] != '\n':
            # this means its a number
            num = float(line)
            current = dict()
            for i in range(100):
                current[wavelengths[i]] = num
                line = fp.readline()
                num = float(line)
            reflectances.append(current)
        line = fp.readline()
```

```python
print(len(reflectances))
```

```python
def draw_reflectance(r):
    x, y = [], []

    for xx, yy in r.items():
        x.append(xx)
        y.append(yy)

    plt.plot(x, y)
    plt.ylim(top=1, bottom=0)
    plt.xlabel('wavelength, nm')
    plt.ylabel('spectral reflectance')
    plt.show()
```

```python
def draw_colour(r):
    dist = colour.SpectralDistribution(r)
    spd = colour.SpectralDistribution(dist)
    csd = clr.sd_to_XYZ(spd)
    csd = colour.XYZ_to_sRGB(csd / 100)
    for k, v in enumerate(csd):
        if v < 0:
            csd[k] = 0
        if v > 1:
            csd[k] = 1
    clplt.plot_single_colour_swatch(clplt.ColourSwatch(RGB=csd))

draw_colour(reflectances[10])
```

```python
import numpy as np
from sklearn.decomposition import PCA

numpyarrays = []
for ref in reflectances:
    t = []
    for k, v in ref.items():
        t.append(v)
    numpyarrays.append(np.array(np.float64(t)))
print(numpyarrays[0])
```

```python
print(numpyarrays[0].shape)
```

```python
pca = PCA(n_components=None)
```

```python
X_std = pca.fit_transform(numpyarrays) 
```

```python
X_std
```

```python
print(X_std[0])
```

```python
var = pca.explained_variance_ratio_
```

```python
print(var[:10])
```

```python
print(np.cumsum(var))
```

Видно, что первые шесть компонент описывают данные на 99%, а значит они лучше всего описывают данные. Получается, что 100-мерное пространство мы можем хорошо приблизить(?) шестимерным.

```python
plt.figure(figsize=(10, 8))
plt.plot(np.linspace(0, 100, 100), var, label='Относительная дисперсия')
plt.plot(np.linspace(0, 100, 100), np.cumsum(var), label='Накопленная относительная дисперсия')
plt.xlabel('Количество главных компонент')
plt.ylabel('Относительное значение')
plt.legend()
plt.show()
```

```python
def draw_illum(r):
    x, y = [], []

    for xx, yy in r.items():
        x.append(xx)
        y.append(yy)

    plt.plot(x, y)
    plt.ylim(top=1, bottom=0)
    plt.xlabel('wavelength, nm')
    plt.ylabel('illum spectrum')
    plt.show()
```

```python
def draw_illums(r):
    fig, ax = plt.subplots(ncols=len(r), nrows=1, figsize=(20, 10))
    j = 0
    for i in range(len(r)):
        x = []
        y = []
        for xx, yy in r[i].items():
            x.append(xx)
            y.append(yy)
        ax[i].plot(x, y)
        ax[i].set_ylim(top=1, bottom=0)
        ax[i].set_title('Illuminant #' +str(i)+ ' of given subset')
```

```python
pca_illum = PCA(n_components=None)
```

```python
def convert_to_np_array(illum):
    r = []
    for i in illum:
        t = np.array([])
        for k, v in i.items():
            t = np.append(t, v)
        r.append(t)   
    return r
```

```python
d = convert_to_np_array(illum)
```

```python
d_transformed = pca_illum.fit_transform(d)
```
